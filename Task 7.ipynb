```python
import matplotlib.pyplot as plt
x = [ 'saturday' , 'sunday' , 'monday' , 'tuesday' , 'wednesday' , 'thursday' , 'friday' ]
y = [ 22  , 24 , 24 , 27 , 29 , 31 , 32 ]

plt.plot( x , y , marker = 'o' , linestyle = '-' , color = 'yellow' )
plt.title( 'Temperature for these days' )
plt.xlabel ( 'Days' )
plt.ylabel ( 'Tempearture' )
plt.xticks(rotation=45)
plt.show()
```
